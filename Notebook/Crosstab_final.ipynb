```python
import pandas as pd

# Memuat data
# Pastikan df sudah ada sebelum menggunakan program, misalnya df = pd.read_csv('data.csv')

# List untuk menyimpan hasil analisis dari setiap iterasi
all_results = []

# Loop agar program berjalan terus-menerus hingga pengguna mengetik 'exit'
while True:
    print("\n=== Analisis Crosstab ===")

    # Input kolom indeks untuk crosstab
    index_column = input("Masukkan kolom indeks (contoh: Q7) atau ketik 'exit' untuk keluar: ").strip().upper()
    if index_column == 'EXIT':
        print("Program dihentikan.")
        break

    # Validasi kolom indeks
    if index_column not in df.columns:
        print(f"Kolom indeks '{index_column}' tidak ditemukan di data. Silakan coba lagi.")
        continue

    # Input kolom yang ingin diproses
    columns_to_analyze = input("Masukkan kolom yang ingin diproses (pisahkan dengan koma): ").split(',')
    columns_to_analyze = [col.strip().upper() for col in columns_to_analyze]

    # Validasi kolom yang diproses
    columns_to_analyze = [col for col in columns_to_analyze if col in df.columns]
    if not columns_to_analyze:
        print("Tidak ada kolom yang valid untuk diproses. Pastikan kolom yang dimasukkan ada dalam data.")
        continue

    # Input kolom values dan jenis agregasi
    value_column = input("Masukkan kolom untuk values (atau ketik 'count' untuk menghitung frekuensi): ").strip()
    if value_column != 'count' and value_column not in df.columns:
        print(f"Kolom '{value_column}' tidak ditemukan di data. Silakan coba lagi.")
        continue

    agg_func = input("Masukkan tipe agregasi (contoh: sum, mean, min, max, count): ").strip().lower()

    # Input metode perhitungan persentase (baris atau kolom)
    percentage_method = input("Masukkan metode persentase (baris / kolom): ").strip().lower()
    if percentage_method not in ['baris', 'kolom']:
        print("Metode persentase tidak valid. Pilih 'baris' atau 'kolom'.")
        continue

    # Dictionary untuk menyimpan hasil crosstab setiap kolom
    crosstab_results = {}

    for col in columns_to_analyze:
        # Tentukan kolom values dan fungsi agregasi
        if value_column == 'count':
            crosstab_result = pd.crosstab(index=df[index_column], columns=df[col])
        else:
            crosstab_result = pd.crosstab(index=df[index_column], columns=df[col], values=df[value_column], aggfunc=agg_func)

        # Hitung persentase berdasarkan metode input
        if percentage_method == 'baris':
            crosstab_percentage = crosstab_result.div(crosstab_result.sum(axis=1), axis=0) * 100
        elif percentage_method == 'kolom':
            crosstab_percentage = crosstab_result.div(crosstab_result.sum(axis=0), axis=1) * 100

        # Konversi hasil crosstab ke bentuk DataFrame
        crosstab_df = crosstab_result.stack().reset_index()
        crosstab_df.columns = [index_column, col, 'frekuensi']

        # Tambahkan kolom persentase
        crosstab_df['persentase'] = crosstab_percentage.stack().values

        # Urutkan data
        crosstab_df = crosstab_df.sort_values([index_column, col]).reset_index(drop=True)

        # Modifikasi untuk format yang diinginkan
        crosstab_df['Kolom yang Dianalisis'] = col
        crosstab_df['value'] = crosstab_df[col]

        # Tambahkan kolom baru untuk menyimpan nama indeks yang digunakan
        crosstab_df['Indeks Yang Digunakan'] = index_column
        # Modifikasi setelah hasil crosstab_df terbentuk
        crosstab_df = crosstab_df.rename(columns={index_column: 'Value_indeks'})

        # Menyusun ulang kolom agar sesuai dengan format
        crosstab_df = crosstab_df[['Kolom yang Dianalisis', 'Indeks Yang Digunakan', 'value', 'Value_indeks', 'persentase']]

        # Batasi desimal menjadi 2 angka
        crosstab_df = crosstab_df.round(3)

        # Simpan hasil crosstab di dictionary
        crosstab_results[col] = crosstab_df

    # Satukan dalam 1 DataFrame dan tambahkan ke list all_results
    crosstab_all = pd.concat(crosstab_results.values(), ignore_index=True)
    all_results.append(crosstab_all)

    # Tampilkan hasil akhir untuk iterasi saat ini
    print("\nHasil Crosstab untuk iterasi ini:")
    print(crosstab_all)

    # Tanya apakah ingin melanjutkan atau keluar
    lanjutkan = input("\nIngin melakukan analisis lagi? (ketik 'exit' untuk keluar atau tekan Enter untuk melanjutkan): ").strip().lower()
    if lanjutkan == 'exit':
        print("Program dihentikan.")
        break

# Gabungkan semua hasil yang telah disimpan dan tampilkan
final_result = pd.concat(all_results, ignore_index=True)
print("\nHasil Gabungan Semua Iterasi:")
print(final_result)

# Opsi untuk mengunduh hasil gabungan sebagai file Excel
download_final_option = input("\nApakah Anda ingin mengunduh hasil gabungan sebagai file excel? (ya/tidak): ").strip().lower()
if download_final_option == 'ya':
    final_file_name = input("Masukkan nama file untuk hasil gabungan (contoh: hasil_gabungan.xlsx): ").strip()
    try:
        final_result.to_excel(final_file_name, index=False)
        print(f"Hasil gabungan telah disimpan sebagai file '{final_file_name}'.")
    except Exception as e:
        print(f"Terjadi kesalahan saat menyimpan file: {e}")
```
